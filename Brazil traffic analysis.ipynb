# Analysing Feasutres Columns

```python
traffic.columns
```

```python
incidents = traffic.drop(["Hour (Coded)" , "Slowness in traffic (%)"] , axis = 1)
```

```python
incidents
```

```python
incidents = incidents.sum().sort_values(ascending = False)
incidents
```

```python
plt.bar(incidents.index,incidents.values)
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

### Hypothesis: 
    * Broke trucks is the max number of incidents

## Checkinng Co-relation with Slowness traffic 

```python
co_relation = traffic.corr()["Slowness in traffic (%)"].sort_values(ascending = False)
co_relation
```

```python
plt.bar(co_relation.index, co_relation.values)
plt.grid()
plt.xticks(rotation = 90)
plt.show()
```

### Result
#### Since first 2 Columns are not Features columns, Therefore, Lack of electricty has the max impact i-e 26% (analysed in target column) on slowness in traffic.
